```python
import tensorflow as tf
from tensorflow import keras
from keras.utils import img_to_array
from keras.utils import load_img
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generate_images(model, img_path):
    test_input = load_img(img_path, target_size=(256,256))
    test_input = img_to_array(test_input).reshape(1,256,256,3)
    test_input = (test_input-127.5)/127.5
    prediction = model(test_input, training=True)
    plt.figure(figsize=(15, 15))
    display_list = [test_input[0], prediction[0]]
    title = ['Input Image', 'Predicted Image']
    for i in range(2):
        plt.subplot(1, 2, i+1)
        plt.title(title[i])
        plt.imshow(display_list[i] * 0.5 + 0.5)
        plt.axis('off')
    plt.show()
```

```python
generator = keras.models.load_model('Models/generator_model .h5')
```

```python
generate_images(generator, 'Data/Custom_Sketches/test1.jpg')
```

```python
generate_images(generator, 'Data/Custom_Sketches/test2.jpg')
```
